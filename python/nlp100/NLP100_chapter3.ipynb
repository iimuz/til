## 20. JSONデータの読み込み

> Wikipedia記事のJSONファイルを読み込み，「イギリス」に関する記事本文を表示せよ．問題21-29では，ここで抽出した記事本文に対して実行せよ．

```
from IPython.display import display
import pandas as pd


def problem020() -> str:
    df = pd.read_json(FILENAME, lines=True)

    uk_text = df[df['title'] == 'イギリス']['text'].values[0]
    print(uk_text)

    return uk_text


uk_text = problem020()
```

## 21. カテゴリ名を含む行を抽出

> 記事中でカテゴリ名を宣言している行を抽出せよ．

```
from pprint import pprint
import re

def problem021() -> None:
    buff = re.findall('.*Category.*', uk_text)

    pprint(buff)


problem021()
```

## 22. カテゴリ名の抽出

> 記事のカテゴリ名を（行単位ではなく名前で）抽出せよ．

```
import re


def problem022() -> None:
    buff = re.findall('.*Category:(.*?)[|\]]+', uk_text)

    print('\n'.join(buff))


problem022()
```

## 23. セクション構造

> 記事中に含まれるセクション名とそのレベル（例えば"== セクション名 =="なら1）を表示せよ．

```
from pprint import pprint
import re


def problem023() -> None:
    sections = {
        iter.group(2): len(iter.group(1)) - 1
        for iter in re.finditer('(={2,})\s*(.+?)\s*=+.*', uk_text)
    }
    pprint(sections)


problem023()
```

## 24. ファイル参照の抽出

> 記事から参照されているメディアファイルをすべて抜き出せ．

```
from pprint import pprint
import re


def problem024() -> None:
    buff = re.findall('\[\[(?:ファイル|File):\s*(.*?)\|.*', uk_text)
    pprint(buff)


problem024()
```

## 25. テンプレートの抽出

> 記事中に含まれる「基礎情報」テンプレートのフィールド名と値を抽出し，辞書オブジェクトとして格納せよ．

```
from pprint import pprint
import re
from typing import Dict

def extract_basic_info(text: str) -> Dict[str, str]:
    """ 基礎情報をフィールド名をキーとする辞書オブジェクトとして抽出する。
    """
    buff = re.findall('\{\{基礎情報.*^\}\}', text, re.MULTILINE | re.DOTALL)
    data = {
        it.group(1): it.group(2)
        for it in  re.finditer(
            '^\|(.+?)\s*=\s*(.+?)(?:(?=\n\|)|(?=\n$))',
            buff[0],
            re.MULTILINE | re.DOTALL
        )
    }

    return data


def problem025() -> None:
    data = extract_basic_info(uk_text)

    pprint(data)


problem025()
```

## 26. 強調マークアップの除去

> 25の処理時に，テンプレートの値からMediaWikiの強調マークアップ（弱い強調，強調，強い強調のすべて）を除去してテキストに変換せよ（参考: マークアップ早見表）．

```
from pprint import pprint
import re

def remove_strong_markup(template: Dict[str, str]) -> Dict[str, str]:
    data = { key: re.sub('\'', '', val) for key, val in template.items() }
    return data


def problem026() -> None:
    data = extract_basic_info(uk_text)
    data = remove_strong_markup(data)

    pprint(data)


problem026()
```

## 27. 内部リンクの除去

> 26の処理に加えて，テンプレートの値からMediaWikiの内部リンクマークアップを除去し，テキストに変換せよ（参考: マークアップ早見表）．

```
from pprint import pprint
import re


def remove_inner_link(src: Dict[str, str]) -> Dict[str, str]:
    p1 = re.compile('\[\[([^:]+?)((\|.*\]\])|(\]\]))')
    dst = { key: re.sub(p1, '\g<1>', val) for key, val in src.items() }
    return dst


def problem027() -> None:
    data = extract_basic_info(uk_text)
    data = remove_strong_markup(data)
    data = remove_inner_link(data)

    pprint(data)


problem027()
```

単純に `[[hoge]]` でかっこ内を置き換えると、ファイルリンクも内部リンクとして処理してしまっていました。
そこで、今回は、 `ファイル:` となっていることから、 `:` を含まない文字の場合に置き換えることにしています。
この方法では、内部リンクに `:` を含んでいた場合に失敗しますが、今回の範囲では大丈夫そうなので、保留とします。

## 28. MediaWikiマークアップの除去

> 27の処理に加えて，テンプレートの値からMediaWikiマークアップを可能な限り除去し，国の基本情報を整形せよ．

```
from pprint import pprint
import re


def remove_file_link(src: Dict[str, str]) -> Dict[str, str]:
    p = re.compile('\[\[ファイル:(.*?)\|(.*?)\|(.*)\]\]')
    dst = { key: re.sub(p, '\g<1>', val) for key, val in src.items() }
    return dst


def remove_outer_link(src: Dict[str, str]) -> Dict[str, str]:
    p = re.compile('\[http.*?\s(.*)\]')
    dst = { key: re.sub(p, '\g<1>', val) for key, val in src.items() }
    return dst


def remove_ref(src: Dict[str, str]) -> Dict[str, str]:
    p = re.compile('<ref.*/>')
    dst = { key: re.sub(p, '', val) for key, val in src.items() }

    p = re.compile('<ref.*?>(.*)</ref>')
    dst = { key: re.sub(p, '\g<1>', val) for key, val in dst.items() }
    return dst


def remove_lang(src: Dict[str, str]) -> Dict[str, str]:
    p = re.compile('{{lang\|.*?\|(.*)}}')
    dst = { key: re.sub(p, '\g<1>', val) for key, val in src.items() }
    return dst


def remove_br(src: Dict[str, str]) -> Dict[str, str]:
    p = re.compile('<br\s*/>')
    dst = { key: re.sub(p, '', val) for key, val in src.items() }
    return dst


def remove_markup(src: Dict[str, str]) -> Dict[str, str]:
    func_list = [
        remove_strong_markup,
        remove_inner_link,
        remove_file_link,
        remove_outer_link,
        remove_ref,
        remove_lang,
        remove_br,
    ]

    dst = src
    for func in func_list:
        dst = func(dst)

    return dst


def problem028() -> None:
    data = extract_basic_info(uk_text)
    data = remove_markup(data)

    pprint(data)


problem028()
```

## 29. 国旗画像のURLを取得する

> テンプレートの内容を利用し，国旗画像のURLを取得せよ．（ヒント: [MediaWiki API][mediawiki]の[imageinfo][imageinfo]を呼び出して，ファイル参照をURLに変換すればよい）

[mediawiki]: http://www.mediawiki.org/wiki/API:Main_page/ja
[imageinfo]: http://www.mediawiki.org/wiki/API:Properties/ja#imageinfo_.2F_ii

```
import json
import urllib


def problem029() -> None:
    data = extract_basic_info(uk_text)
    data = remove_markup(data)

    request_url = (
        'https://www.mediawiki.org/w/api.php?'
        'action=query'
        f'&titles=File:{urllib.parse.quote(data["国旗画像"])}'
        '&format=json'
        '&prop=imageinfo'
        '&iiprop=url'
    )
    with urllib.request.urlopen(request_url) as response:
        res = json.loads(response.read().decode())
    image_url = res['query']['pages']['-1']['imageinfo'][0]['url']

    print(image_url)


problem029()
```
